```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load CSV files into DataFrames
df = pd.read_csv('./data/AAs.csv')
df.head()
```

```python
sns.set(style="whitegrid")
# Assuming 'df' is already defined and processed
df['group'] = df['name'].str[:2]

df_melted = df.melt(id_vars=['name', 'group'], var_name='amino_acid', value_name='content')
df_melted = df_melted[df_melted['amino_acid'] != 'name']

# Create the plot
plt.figure(figsize=(7, 4))
sns.boxplot(data=df_melted, x='amino_acid', y='content', hue='group', showfliers=False)
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
# plt.savefig("amino_acid_boxplot.pdf", format="pdf")
plt.show()
```
